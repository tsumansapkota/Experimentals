```python
# Now I have to 
#1. fully optimize the regression
#2. Calculate the most error region of the convex function (given by eror analysis)
#3. Break the end points and optimize 2 piece network
```

```python
hidden_size = 1

optimizer = tnn.Adam()
layer1 = tnn.NonLinearLayer(1, hidden_size, activation=tnn.Linear(), optimizer=optimizer)
layer2 = tnn.LinearLayer(hidden_size, 1, optimizer=optimizer)
```

```python
%matplotlib tk
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.ion()
fig.show()
fig.canvas.draw()

for step in range(10000):
    a1 = layer1.forward(x_data)
    a2 = layer2.forward(a1)
    loss = tnn.MseLoss.loss(a2, y_data)
    loss_del = tnn.MseLoss.del_loss(a2, y_data)
    
    da1 = layer2.backward(loss_del)
    dx = layer1.backward(da1)
    
    layer1.update()
    layer2.update()
    
    if step % 1 == 0:

        print('_______________________')
        print("epoch", step)
        print("Loss", loss)

        ax.clear()
        plt.scatter(x, y, marker='.', c='g', label='x,y')
        plt.scatter(x, a2.reshape(-1), marker='.', c='r', lw=3, label='x,a2')
        plt.scatter(x, dx.reshape(-1), marker='.', c='c', lw=1, label='x,dx')
        plt.plot(x, np.zeros_like(x), c='b', lw=2)
#         plt.scatter(x, da1.reshape(-1), marker='.', c='y', lw=1, label='x,da1')
        plt.legend()
        
        fig.canvas.draw()
        
        print("Weight 1", '\n', layer1.weights)
        print("Biases 1", '\n', layer1.bias)
        print("Weight 2", '\n', layer2.weights)
        print("Biases 2", '\n', layer2.bias)
        
        plt.pause(0.001)
        print("\n")
        if loss < 0.0025:
            print('fitted to our expectation')
            plt.pause(50)
            break
```
