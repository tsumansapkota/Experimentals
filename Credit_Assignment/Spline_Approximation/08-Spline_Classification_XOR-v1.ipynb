```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import mylibrary.nnlib as tnn
import matplotlib
import splinelib as tsp
```

```python
T = 1.
F = 0.

xxx = np.array([[F, F],
      [F, T],
      [T, F],
      [T, T],])
yyy = np.array([[F],[T],[T],[F]])

x = []
y = []
n_data = 200
for i in range(n_data):
    index = np.random.randint(low=0, high=4)
    xx = xxx[index] + np.random.normal(loc=0., scale=0.15, size=xxx[0].size)
    x.append(xx)
    y.append(yyy[index])
x, y = np.array(x), np.array(y)
print(x.shape, y.shape)
plt.scatter(x[:,0], x[:, 1], c=y.reshape(-1))
```

```python
colors = ['r', 'y', 'g', 'b']
```

```python
lr=0.1
max_points = 5

linear = tnn.LinearLayer(2,1, optimizer=tnn.SGD(lr))
x_ = linear.forward(x)
spline = tsp.Spline1D(max_points, x_.reshape(-1), y)

%matplotlib inline
plt.scatter(x_, y, marker='.', color='g')
plt.plot(spline.X, spline.Y)
```

```python
plt.scatter(x[:,0], x_, marker='+',c=y.reshape(-1))
plt.scatter(x[:,1], x_, marker='_',c=y.reshape(-1))
```

```python
%matplotlib tk
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.ion()
fig.show()
fig.canvas.draw()
y_ = y.reshape(-1)
for step in range(10000):
    x_ = linear.forward(x).reshape(-1)
    Y = spline.forward(x_)
    dy = Y-y_
    dx_ = spline.backward(dy)
    dx = linear.backward(dx_.reshape(-1,1)).reshape(-1)
    
    spline.update(lr)
    linear.update()
    
    if step % 100 == 0:
        spline._remove_close_points_()
        spline._add_new_point_()
        spline._remove_no_input_points_()
        
    if step % 10 == 0:
            
        error = (dy**2).mean()
        
        print('_______________________')
        print("Steps", step)
        print("Error", error)
        print("n_points", spline.n_points)
        
        out = (Y>0.5).astype(int)        
        ax.clear()
        plt.scatter(x_, y, marker='.', color='g')
        plt.plot(spline.X, spline.Y, marker='.', color='r')
        plt.scatter(x_, out, marker='.', color='r')
        

        
        fig.canvas.draw()
        plt.pause(0.01)
        print("\n")
        if error < 0.00025:
            print('fitted to our expectation')
            plt.pause(50)
            break
```
