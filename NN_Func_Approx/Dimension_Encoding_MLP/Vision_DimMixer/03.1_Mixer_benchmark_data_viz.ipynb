```python
import numpy as np
import json
import os, sys, pathlib
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
layers = [7, 10]
DS = 'c10'
SEEDs = [147, 258, 369]
```

```python
## find best test accuracy and number of params.
data_dict = {'model': [], 'acc': [], 'params':[], 'layers':[], 'seed':[]}
for SEED in SEEDs:
    for num_layers in layers:
        model_names = [f'original_mixer0',
                        f'original_mixer1',
                        f'patchonly_mixer0' ]
        for model_name in model_names:
            ## load stat of the model
#             print(model_name)
            file = f'./output/03.0_{model_name}_l{num_layers}_{DS}_s{SEED}_data.json'
            if not os.path.exists(file):
                print(f"File: {file} not found. Next ...")
                continue
                
            with open(file, 'r') as f:
                data = json.load(f)    
            
            best_acc = np.array(data['test_stat'])[:, 2].max()
            if data['test_stat'][-1][0] < 199:
                print(f"Incomplete training {file} {data['test_stat'][-1][0]}")
                continue
                
#             print(best_acc)
            data_dict['model'] += [model_name]
            data_dict['acc'] += [best_acc]
            data_dict['params'] += [data['num_params']]
            data_dict['layers'] += [num_layers]
            data_dict['seed'] += [SEED]
        pass
    pass
pass
```

```python
df = pd.DataFrame(data_dict)
```
