```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
T = 1.
F = -1.
```

```python
inputs = np.array([[T, T ,1], # X1, X2, b
                  [T, F ,1],
                  [F, T ,1],
                  [F, F ,1]])
print(inputs)
print(inputs.shape)
```

```python
outputs = np.array([[T],
                   [F],
                   [F],
                   [F]])
print(outputs)
print(outputs.shape)
```

```python
Y = outputs
X = inputs
```

```python
W = np.random.randn(3,1)
W
###
# Basic initialization done
```

```python
# trying the traditional method
y = X.dot(W)
delta = y - Y
dW = X.T.dot(delta) #* 1./delta.shape[0] ## 1./m
alpha = 0.1
W -= alpha * dW
# the new output after the update
y = X.dot(W)
y
```

```python
# loop begins here
Wi = np.linalg.pinv(W)
Wi
```

```python
y = X.dot(W)
print(y)
print(y.shape)
```

```python
x = y.dot(Wi)
x
```

```python
delta = Y - y
print(delta)
delta = delta**2
delta.mean()
```

```python
defta = X - x
print(defta)
defta = defta**2
defta.mean()
```

```python
dW1 = X.T.dot(delta)/delta.shape[0]
dW1
```

```python
W = W - dW1 * 0.1
W
```

```python
# after forward encoding correction:
y = X.dot(W)
delta = Y - y
print(delta)
delta = delta**2
delta.mean()
```

```python
dWi2 = Y.T.dot(defta)/defta.shape[0]
dWi2
```

```python
Wi = Wi - dWi2 * 0.1
Wi
```

```python
Wii = np.linalg.pinv(Wi)
Wii
```

```python
# after backward encoding correction:
y = X.dot(Wii)
delta = Y - y
print(delta)
delta = delta**2
delta.mean()
```

```python
W = (W + Wii)/2
W
```

```python
delta = Y - X.dot(W)
print(delta)
delta = delta **2
delta.mean()
```

```python
for i in range(0, 100):
    print("Weight = ", W)
    Wi = np.linalg.pinv(W)
    print("Weight I ", Wi)
    
    y = X.dot(W)
    print("X.W = ", y)
    x = y.dot(Wi)
    print("y.Wi = ", x)
    delf = Y - y
    print("delf = ", delf, (delf**2).mean())
    delb = X - x
    print("delb = ", delb, (delb**2).mean())

    dW = X.T.dot(delf)/delf.shape[0]
    dWi = Y.T.dot(delb)/delb.shape[0]
    W = W - dW * 0.01
    Wi = Wi - dWi * 0.01
    print('Wi ' , Wi)

    Wii = np.linalg.pinv(Wi)
    print('Wii ' , Wii)
    W = (W + Wii)/2
    print()
    print("New W = ", W)
    print("new del = ", Y - X.dot(W))
```

```python
T = 1
F = 0
X = np.array([[T, T],
                  [T, F],
                  [F, T],
                  [F, F]])
Y = np.array([[T],
                   [F],
                   [F],
                   [F]])
```

```python
W = np.random.randn(2,1)
W
```

```python
for i in range(0, 100):
    print("Weight = ", W)
    
    y = X.dot(W)
    print("X.W = ", y)

    delf = Y - y
    print("delf = ", delf)

    dW = X.T.dot(delf)/delf.shape[0]
    W = W - dW * 0.1
    delf = Y - X.dot(W)
    print("new del = ", delf)
    print(delf.mean())
    print()
```
