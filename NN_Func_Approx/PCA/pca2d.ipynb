```python
inp  = np.c_[x,y]
```

```python
# REAL action down here
W = np.random.uniform(-0.7, 0.7, size=(2,1))
# W_ = np.linalg.pinv(W)
W_ = W.T
learning_rate = 0.1

for i in range(1000):
    h = inp.dot(W)
    inp_ = h.dot(W_)
    
    dE = inp_- inp
    error = np.mean(dE**2)
    print('error = ',error)
    
    dh = dE.dot(W_.T) * h
    dW = x.T.dot(dh)
    W -= learning_rate * dW
#     W_ = np.linalg.pinv(W)
    W_ = W.T
    print('Weights', W)
```
