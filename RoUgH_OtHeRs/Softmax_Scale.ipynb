```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arr = torch.rand(5)*2-1
arr
```

```python
sft1 = torch.softmax(arr, dim=0)
sft1
```

```python
sft64 = torch.softmax(64*arr, dim=0)
sft64
```

```python
plt.plot(arr, label='arr')
plt.plot(sft1, label='s_1')
plt.plot(sft64, label='s_64')
plt.legend()
```
