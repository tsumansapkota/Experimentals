```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import mylibrary.nnlib as tnn
```

```python
x = np.linspace(0,1,300)
noise = np.random.normal(0,0.02, x.shape)
x = x + noise
y = (2 * np.sin(10*x) + np.cos(20 * x - 3) + 3 * np.log(10*x + 0.5) - 4)/6.
noise = np.random.normal(0,0.1, x.shape)
y = y + noise
x = x - 0.4 #x.mean() # good practice to zero-center x in linear mapping
sortindx = np.argsort(x)
x = x[sortindx]
y = y[sortindx]
xx = x.reshape(-1,1)
yy = y.reshape(-1,1)
plt.scatter(x, y, marker='.', color='g')
```

```python
optimizer=tnn.Adam(0.03)
hid1=6

layer1 = tnn.LinearLayer(1,hid1)
layer2 = tnn.LeakyRelu()
# layer2 = tnn.Relu()
layer3 = tnn.LinearLayer(hid1,1)

net = tnn.AutoForm()
net.collect_global_layers()
```

```python
%matplotlib tk
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.ion()
fig.show()
fig.canvas.draw()

for step in range(100000):

    a2 = net.forward(xx)
    error = tnn.MseLoss.loss(a2,yy)
    dy = tnn.MseLoss.del_loss(a2,yy)
    
    dx = net.backward(dy)
    net.update()
    
    if step % 100 == 0:            
        print('_______________________')
        print("Steps", step)
        print("Error", error)
        
        ax.clear()
        plt.scatter(x, y, marker='.', color='g')
        plt.scatter(x, a2, marker='.', color='r')
        plt.scatter(x, dx, marker='.', color='c')
        
        fig.canvas.draw()
        plt.pause(0.01)
        print("\n")
        if error < 0.00025:
            print('fitted to our expectation')
            plt.pause(50)
            break
```

```python
def reverse_leakyRelu(self:tnn.LeakyRelu, y):
    dx = np.ones_like(self.x)
    dx[self.x < 0] = 1/self.alpha
    return dx * y
np.abs(reverse_leakyRelu(layer2, layer3.input) - layer2.x).mean()
```
